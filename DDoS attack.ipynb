```python
def refactoringParameters():
    categoricalParameters = []
    
    protocol = list(set(df['protocol_type'].values.tolist()))
    proIndex = range(0,len(protocol))
    
    
    protocolType =  dict(zip( protocol,proIndex))
    df['protocol_type'] =  df['protocol_type'].map(protocolType)
    categoricalParameters.append(protocolType)
    ff = list(set(df['flag'].values.tolist()))
    ffIndex = range(0,len(ff))
    
    flagInfo =  dict(zip( ff,ffIndex))
    df['flag'] =  df['flag'].map(flagInfo)
    categoricalParameters.append(flagInfo)
    service = list(set(df['service'].values.tolist()))
    serviceIndex =  range(0,len(service))
    serviceInfo = dict(zip(service,serviceIndex))
    df['service'] =  df['service'].map(serviceInfo)
    categoricalParameters.append(serviceInfo)
#     labelOut = list(set(df['label'].values.tolist()))
#     labelOutIndex =  range(0,len(labelOut))
#     labelOutInfo = dict(zip(labelOut,labelOutIndex))
#     df['label'] =  df['label'].map(labelOutInfo)
#     categoricalParameters.append(labelOutInfo)
    return categoricalParameters


if __name__ == '__main__':
    categoricalParameters = refactoringParameters()
    print("Categorical Parameters:", categoricalParameters)
    filename = 'Category/DDOS_categoricalValues.sav'
    pickle.dump(categoricalParameters, open(filename, 'wb'))
    df.head()
```

```python
# Creating a pickle models from datasets to classify the ddos attacks
trafficData = pd.read_csv('dataset/traffic.csv')

set(trafficData['label'].values)
# Deleting Unwanted Column
trafficData =  trafficData.drop(['Unnamed: 0'], axis=1)
trafficData

```

```python
# Changing label names to numbers

labelOut = list(set(trafficData['label'].values.tolist()))
labelOutIndex =  range(0,len(labelOut))
labelOutInfo = dict(zip(labelOut,labelOutIndex))
trafficData['label'] =  trafficData['label'].map(labelOutInfo)
trafficData.head()
```

```python
# GFG random forest model generations, convert gfg_blog_data.csv to sav model
def generateRandomForestModel(X_train, y_train,filename):
    RF = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 42)  
    RF.fit(X_train, y_train)
    filename = 'ML_model/'+filename+'.sav'
    pickle.dump(RF, open(filename, 'wb'))
    return RF

def datasetParameters(Data):
    predict_label = Data.label
    X = Data.drop('label', axis=1)
    X_train, X_test, y_train, y_test = train_test_split(X, predict_label ,test_size=0.2)
    totalData = X_train, X_test, y_train, y_test
    return totalData
```

```python
X_train, X_test, y_train, y_test = datasetParameters(trafficData)

VV = generateRandomForestModel(X_train, y_train,'traffic_rf_model')
print('Model Created Successfully for Traffic dataset using Random Forest')
```
